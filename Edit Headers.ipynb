## Shots Tables

```python
# Shots tables
large_df = pd.read_csv("data/NBA_Shots_2000_to_2018.csv", index_col=0)
large_df.head()
```

```python
small_df = pd.read_csv("data/samples/sample_NBA_Shots_2000_to_2018.csv", index_col=0)
small_df.head()
```

```python
cols = {col: col.lower().replace(" ", "_").replace(".", "_").replace("(", "").replace(")", "").replace(",", "").replace("__", "_").replace("t_", "t") for col in large_df.columns}
cols
```

```python
large_df.rename(columns=cols, inplace=True)
small_df.rename(columns=cols, inplace=True)
large_df.to_csv("data/NBA_Shots_2000_to_2018.csv", index=False)
small_df.to_csv("data/samples/sample_NBA_Shots_2000_to_2018.csv", index=False)
```
